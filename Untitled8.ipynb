```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from  sklearn.metrics import silhouette_score
```

```python
data=pd.read_csv('final_dataset (1).csv')
data.head()
```

```python
scaler=StandardScaler()
data_scaled=scaler.fit_transform(data)
print(pd.DataFrame(data_scaled,columns=data.columns).head())
```

```python
interia=[]
k_range=range(1,11)
for k in k_range:
     kmeans=KMeans(n_clusters=k,random_state=42)
     kmeans.fit(data_scaled)
     interia.append(kmeans.inertia_)
plt.plot(k_range,interia,marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Interia')
plt.show()
```

```python
k=2
kmeans=KMeans(n_clusters=k,random_state=52)
data['cluster']=kmeans.fit_predict(data_scaled)
data.head()
```

```python
sil_score=silhouette_score(data_scaled,data['cluster'])
print(f'silhoutette score:{sil_score}')
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x='NO2',y='SO2',hue='cluster',data=data,palette='dark',s=200,alpha=0.8,edgecolor='g')
plt.title('Cluster data')
plt.xlabel('NO2')
plt.ylabel('AQI')
plt.legend(title='cluster',bbox_to_anchor=(1.05,1),loc='upper left')
plt.show()
```
