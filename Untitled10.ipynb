```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('green transportation system.csv')
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data_scaled=scaler.fit_transform(data)
scaler_data=pd.DataFrame(data_scaled,columns=data.columns)
scaler_data.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
pca.fit(data_scaled)
print("Explained variance ratio for each pripncipal component:")
print(pca.explained_variance_ratio_)
print("Cumulative explained variance ratio:")
print(np.cumsum(pca.explained_variance_ratio_))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.bar(range(1,len(pca.explained_variance_ratio_)+1),pca.explained_variance_ratio_)
plt.title('Explained Variance by Each pripncipal component')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1,len(pca.explained_variance_ratio_)+1),np.cumsum(pca.explained_variance_ratio_),marker='o',linestyle='--')
plt.title('Cumulative Explained Variance by Each pripncipal component')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.show()
```

```python
data_pca=pca.transform(data_scaled)
pca_df=pd.DataFrame(data_pca[:,:2],columns=['PC1','PC2'])
plt.figure(figsize=(8,6))
plt.scatter(pca_df['PC1'],pca_df['PC2'],c='red',edgecolors='k',alpha=0.7)
plt.title('PCA of Green Transportation System')
plt.xlabel('PrincipalComponent1')
plt.ylabel('PrincipalComponent2')
plt.show()
```
