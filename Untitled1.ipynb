```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import joblib
```

```python
df=pd.read_csv('/content/solar_power_output.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
X=df[['temperature','humidity','solar_irradiance','wind_speed']]
y=df['solar_power_output']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
mse= mean_squared_error(y_test,y_pred)
print(f"Mean squard Error:{mse}")
r2score=r2_score(y_test,y_pred)
print(f"R2 score:{r2score}")
```

```python
plt.figure(figsize=(10,5))
plt.scatter(y_test,y_pred,color='g',label='Actual vs predicted')
plt.plot([y.min(),y.max()],[y.min(),y.max()],
         color='purple',linestyle='--',label='Best fit line')
plt.xlabel('Actutal Solar power output')
plt.ylabel('predicted solar power output')
plt.title('Actual vs predicted solar power output')
plt.show()
```
