```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
num_samples=100
```

```python
carbon_emissions=np.random.uniform(50,400,num_samples)
energy_output=np.random.uniform(100,1000,num_samples)
renewability_index=np.random.uniform(0,1,num_samples)
cost_effieciency=np.random.uniform(0.5,5,num_samples)
```

```python
sustainability=[1 if (emission<200 and renewability>0.5 and cost <3) else 0
                for emission,renewability,cost in zip(carbon_emissions,renewability_index,cost_effieciency)]
```

```python
data=pd.DataFrame({
    'carbon_emissions':carbon_emissions,
    'energy_output':energy_output,
    'renewability_index':renewability_index,
    'cost_effieciency':cost_effieciency,
    'sustainability':sustainability
})
data
```

```python
data.to_csv('sustainability_datas.csv',index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('sustainability_datas.csv')
print(data.head())
```

```python
x=data [['carbon_emissions','energy_output','renewability_index','cost_effieciency']]
y=data['sustainability']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model= LogisticRegression()
model.fit(x_train,y_train)
```

```python
   y_pred=model.predict( x_test)
```

```python
y_pred
```

```python
#accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
conf_matrix = confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(conf_matrix,annot=True,fmt='d',cmap='Blues',xticklabels=['Not Sustainable','Sustainable'] ,yticklabels=['Not Sustainable','Sustainable'])
plt.xlabel('predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
